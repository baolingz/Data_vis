```python
import coffee_script as cs
%matplotlib inline
```

```python
# cities to work with
cities_to_states = {'New York':'NY',
                    'Los Angeles': 'CA', 
                    'Chicago':'IL',
                    'Dallas': 'TX',
                    'Houston':'TX',
                    'Washington':'DC',
                    'Philadelphia': 'PA',
                    'Miami': 'FL',
                    'Atlanta': 'GA',
                    'Boston':'MA', 
                    'San Francisco':'CA',
                    'Detroit': 'MI',
                    'Denver': 'CO',
                    'Charlottesville': 'VA',
                    'Baltimore': 'MD',
                    'Charlotte': 'NC',
                    'Pittsburgh': 'PA',
                    'Austin': 'TX',
                    'Cleveland': 'OH',
                    'Columbus': 'OH',
                    'Portland': 'OR',
                    'New Orleans': 'LA',
                    'Salt Lake City': 'UT',
                    'Seattle':'WA'}

cities = sorted(list(cities_to_states.keys()))
```

```python
mega_map, mega_data = cs.megalist(cities)
```

```python
mega_map.plot(figsize=(16,16))
```

```python
if 'Cambridge' not in cities:
    cities.insert(4, 'Cambridge')
for city in cities[0:1]:
    city_map  = mega_map[mega_map.City == city]
    city_data = mega_data[mega_data.City == city]
    cs.plot_map(city_map, city_data, 'count', city=city)
```
